```python
# import packages 
import time
import pandas as pd
import numpy as np
import os
import copy
os.chdir("C:\\Users\\omkar\\Documents\\OM\\Bike_Share\\Datasets")
# Load the dataset by global variable 
CITY_DATA = { 'ch': 'chicago.csv',
              'ny': 'new_york_city.csv',
              'w': 'washington.csv' }
```

```python
def get_filters():
    print('Hello! Let\'s explore some US bikeshare data!\n')
    while True:
        city = input("please, pick a city to analyze:(ch) for chicago or (ny) for new_york_city or (w) for washington:\n").lower()
        if city in CITY_DATA.keys():
            break
        else:
            print("Opps!! Invalid Input...")
    months = ["jan", "feb", "mar", "apr", "may", "jun", "all"]
    while True:
        month=input("please, pick month (jan, feb, mar, apr, may, jun) to filter or type (all) for not filtering:\n").lower()
        if month in months:
            break
        else:
            print("Opps!! Invalid Input...")
    days = ["sat", "sun", "mon", "tue", "wed", "thu", "fri", "all"]
    while True:
        day = input("please, pick day of week(sat, sun, mon, tue, wed, thu, fri)to filter or(all) for not filtering:\n").lower()
        if day in days:
            break
        else:
            print("Opps!! Invalid Input...")
    print('-'*40)
    return city,month,day

def load_data(city,month,day):
    df=pd.read_csv(CITY_DATA[city])
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['month']=df['Start Time'].dt.month_name()
    df['day of week']=df['Start Time'].dt.day_name()
     # filter by month if applicable
    if month != 'all':
        # filter by month to create the new dataframe
        df = df[df['month'].str.startswith(month.title())]

    # filter by day of week if applicable
    if day != 'all':
        # filter by day of week to create the new dataframe
        df = df[df['day of week'].str.startswith(day.title())]
    return df
def time_stats(df):
    most_common_month=df['month'].mode()[0]
    print(f'Most common month:{most_common_month}')
    most_common_day=df['day of week'].mode()[0]
    print(f'Most common day:{most_common_day}')
    df['hour'] = df['Start Time'].dt.hour
    most_common_hour=df['hour'].mode()[0]
    if most_common_hour<12:
        print('\nThe most common hour of travel is: {} AM\n'.format(most_common_hour))
    elif most_common_hour>12 & most_common_hour<24:
        print('\nThe most common hour of travel is: {} PM\n'.format(most_common_hour-12))
    elif most_common_hour==24:
        print('\nThe most common hour of travel is: {} AM\n'.format(most_common_hour-12))
    print('-'*40)
def station_stats(df):
    print('\nCalculating The Most Popular Stations and Trip...\n')
    # TO DO: display most commonly used start station
    most_common_start = df['Start Station'].mode()[0]
    print('The most commonly used start station is:', most_common_start) 

    # TO DO: display most commonly used end station
    most_common_end = df['End Station'].mode()[0]
    print('The most commonly used end station is:', most_common_end)

    # TO DO: display most frequent combination of start station and end station trip
    df1 = df.groupby(['Start Station','End Station'])['User Type'].count()
    common_trip=df1[df1==df1.max()].to_frame()
    print('Most frequent combination of start station and end station trip \n', common_trip)
    print('-'*40)

def travel_time(df):
    print('\nCalculating Total trip duration and mean trip duration...\n')
    total_travel_time=df['Trip Duration'].sum()
    print(f'Total travel time is-{total_travel_time}')
    mean_travel_time=df['Trip Duration'].mean()
    print(f'Mean travel time is-{mean_travel_time}')
    print('-'*40)
        
def User_info(df):
    print('\nCalculating User Type,Gender and Birth counts related to Bike_Taxi...\n')
    User_count=df['User Type'].value_counts().to_frame()
    print('Here are users types:\n',User_count)
    try:
        print("Gender is\n", df['Gender'].value_counts().to_frame())
        print('Earliest year of birth is', df['Birth Year'].min())
        print('Most recent of birth is', df['Birth Year'].max())
        print('Most common year of birth is', df['Birth Year'].mode()[0])    
        
    except:
        print('No filter with gender allowed in Washington city!')
    print('-'*40)
    
def ask_more_data(df):
    getmoredata=input('Need first 5 rows /yes or /no:').lower()
    start_loc=5
    while getmoredata=='yes':
        print(df.iloc[0:start_loc])
        start_loc+=5
        getmoredata=input('Ask more data type /yes or /no:').lower()
        
def main():
    
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        travel_time(df)
        User_info(df)
        ask_more_data(df)
        
        return_data=input('Enter if you want the data again /yes or /no:')
        if return_data!='yes':
            break

if __name__ == "__main__":
	main()
```
